```python
###changing the other colums to float data type
```

```python
### Exploratory data analysis
```

```python
df_copy = df.drop(['day','month','year'],axis=1)
```

```python
df_copy.head()
```

```python
### encoding of the categories in classes
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
df_copy.head()
```

```python
df_copy.tail()
```

```python
df_copy['Classes'].value_counts()
```

```python
### plot the density plot for all features
```

```python
# plt.style.use('seaborn')
df_copy.hist(bins =50,figsize=(20,15))
plt.tight_layout()

```

```python
### percentage for pie chart
percentage = df_copy['Classes'].value_counts(normalize=True)*100 ## normalize gives us percentage
```

```python
### ploting pie chart
classlabels = ["Fire","Not Fire"]
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title("Pie chart of Classes")
plt.savefig(r"images/Pie_Chart_Of_Classe.png")
```

```python
### correlation
```

```python
df_copy.corr()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(data=df_copy.corr())
plt.title("Correlation between Features")
plt.tight_layout()
plt.savefig(r"images/Correlation_Plot.png")
```

```python
### box plot
plt.figure(figsize=(12,8))
sns.boxplot(data=df_copy)
plt.tight_layout()
plt.title("Boxplot Between Features")
plt.savefig(r"images/Boxplot_between_features.png")
```

```python
df['Classes'] = np.where(df['Classes'].str.contains('not fire'),'not fire','fire')
```

```python
## monthly fire analysis
dftemp = df.loc[df['Region']==0]
plt.subplots(figsize =(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data =df)
plt.xlabel('Months',weight ='bold')
plt.ylabel("Number Of Fires",weight = 'bold')
plt.title("Fire Analysis of Brjaia Region")
plt.savefig(r"images/Monthly_Fire_Analysis_Brjaia_Region.png")
```
